```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
sns.set_theme()
```

```python
df_train = pd.read_csv(r"C:\Users\Marce\Downloads\house-prices-advanced-regression-techniques\train.csv")
df_test = pd.read_csv(r"C:\Users\Marce\Downloads\house-prices-advanced-regression-techniques\test.csv")
```

```python
result_l = pd.concat([df_train, df_test], ignore_index=True)
result = result_l.drop(['Id'],axis=1)
```

```python
X = result.iloc[:, :-1]
y = result.iloc[:, -1]
```

```python
object_columns = df_train.select_dtypes(include=['object']).columns
onehot_encoder = OneHotEncoder(sparse=False, drop='first')
```

```python
for column in object_columns:
    column_data = X[column].values.reshape(-1, 1)
    
    encoded_data = onehot_encoder.fit_transform(column_data)
    
    new_column_names = [f"{column}_{category}" for category in onehot_encoder.get_feature_names_out([column])]
    
    encoded_df = pd.DataFrame(encoded_data, columns=new_column_names)
    
    X = pd.concat([X, encoded_df], axis=1)
    
    X = X.drop(column, axis=1)
```

```python
train_indices = range(1459)  

X_train, X_test, y_train, y_test = train_test_split(X.loc[train_indices], y.loc[train_indices], test_size=0.25, random_state=60)
```

```python
for column in X_train.columns:
    X_train[column].fillna(0, inplace=True)

for column in X_test.columns:
    X_test[column].fillna(0, inplace=True)
for column in X_test.columns:
    X[column].fillna(0, inplace=True)
```

## Random Classifier Method


```python
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 42)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred_vec = y_pred.reshape(-1,1)
```

```python
y_test_vec = np.array(y_test).reshape(-1,1)
```

## R_Square

```python
concatenated_array  = np.concatenate((y_pred_vec,y_test_vec), axis = 1)
```

```python
from sklearn.metrics import r2_score
r2_value = r2_score(y_test, y_pred)
print("R-squared:", r2_value)
```

```python
variance_per_feature = np.mean(concatenated_array, axis=0)
print(variance_per_feature)
```

```python
X_to_apply = X.iloc[1460:]

y_pred_final = classifier.predict(X_to_apply)
```
