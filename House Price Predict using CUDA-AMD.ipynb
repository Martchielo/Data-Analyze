```python
train_data, test_data, _, _ = train_test_split(df_train, df_train['SalePrice'], test_size=0.3, random_state=0)
```

```python
test_data
```

```python
train_GPU = tfdf.keras.pd_dataframe_to_tf_dataset(train_data, label='SalePrice', task=tfdf.keras.Task.REGRESSION)
```

```python
test_GPU = tfdf.keras.pd_dataframe_to_tf_dataset(test_data, label='SalePrice', task = tfdf.keras.Task.REGRESSION)
```

* Adjust Parameters

```python
model = tfdf.keras.RandomForestModel(task = tfdf.keras.Task.REGRESSION)
```

```python
model.fit(train_GPU)
```

```python
model.compile(metrics = ['mse'])
```

```python
evaluation = model.evaluate(x=test_GPU,return_dict=True)

for name, value in evaluation.items():
  print(f"{name}: {value:f}")
```
